Preprocessing(ConvertToLower,remove Punc and stopwords)

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
s=''
corpus=[]
for i in range (len(data)):
    s=re.sub("[^a-zA-Z]"," ",data["Review"][i])
    s=word_tokenize(s)
    s=[word.lower() for word in s if word not in stopwords.words("english")]
    s=' '.join(s)
    s=lemma.lemmatize(s)
    print(s)
    corpus.append(s)     
print(corpus)    
```

Feature Extraction

```python
cv=CountVectorizer()
cv.fit(corpus)#convert each word into representaion like-->(eat)-->(1 0 0) and Learn vocab
#joblib.dump(cv, 'CV.joblib')
X=cv.transform(corpus).toarray()#convert representaion into vector
print(X)
cv.vocabulary_#position       
```

```python
labels=cv.get_feature_names_out()#to get Vocabulary OR word without position
print(labels)
#pd.DataFrame(data=X,columns=cv.get_feature_names_out())#to set Vocab and repetition of words 
```

```python
y=data["Liked"]
print(y)
```

Train The Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.20,random_state=40)
```

```python
from sklearn.naive_bayes import MultinomialNB
##-->acuuracy=0.81-->>>good
#from sklearn.linear_model import LogisticRegression
#model = LogisticRegression()

#-->accuracy=0.66-->so bad 
#from sklearn.neighbors import KNeighborsClassifier
#model=KNeighborsClassifier(n_neighbors=23)

model=MultinomialNB()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,y_pred))
```
